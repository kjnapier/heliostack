```python
import torch
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

### Choosing the hardware to run on

```python
device = torch.device("cuda")
# device = torch.device("cpu")
```

### Generating a random torch tensor

```python
arr = torch.rand(64, 64, device=device, dtype=torch.half)

# You can slice and mutate tensors like numpy arrays
arr[30:50, 30:50] = 0
arr[34:46, 34:46] = 1
arr[37:43, 37:43] = 0

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(arr.cpu(), cmap='gray')
```

### Instantiating a torch tensor from a numpy array

```python
cpu_arr = np.random.rand(64, 64)
arr = torch.tensor(cpu_arr, device=device, dtype=torch.half)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(arr.cpu(), cmap='gray')
```
